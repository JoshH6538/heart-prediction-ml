# K-Nearest Neighbors Classification - Heart Failure Prediction

This notebook builds and evaluates a **K-Nearest Neighbors Classification** model for predicting the likelihood of heart failure based on patient health attributes.  

---

## Objective

Use K-Nearest Neighbors as a non-parametric classifier to:
- Predict the presence of heart disease based on similarity to other patients
- Compare its performance with models such as Logistic Regression, Decision Trees, and Random Forests
- Explore how the choice of k (number of neighbors) and distance metric affects prediction accuracy and model stability
---

## Dataset Summary

The dataset contains **918 patient records** and **11 health-related features**, along with a binary target variable:

- **Target:** `HeartDisease` (1 = patient has or is at risk of heart failure, 0 = otherwise)

**Features include:**
- *Age, Sex, ChestPainType, RestingBP, Cholesterol, FastingBS, RestingECG, MaxHR, ExerciseAngina, Oldpeak, ST_Slope*

---

## Overview

1. **Load the dataset** (dataset/heart.csv) 
2. **Preprocess features** (encode categorical variables and scale numerical features)  
3. **Train and evaluate** a KNN classifier using both a **standard 80/20 train–test split** and **10-fold cross-validation** 
4. **Compare performance** across the two evaluation methods to assess model stability and generalization  

---

## Notes

- All preprocessing (encoding, scaling, and splitting) is performed **within this notebook** for simplicity and reproducibility.  
- A fixed random seed (`random_state=42`) is used to maintain consistent splits and fair comparisons between methods.  

---


```python
# Gathering all neccessary libraries
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, auc
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import cross_validate
import matplotlib.pyplot as plt
```

## Reading our Dataset

We read our dataset from the Kaggle .csv file and and create our DataFrame

```python
df = pd.read_csv("dataset/heart.csv")

df.head()
```

```python
df.shape
```

```python
df.info()
```

## Encoding our categorical features

Before we create our feature matrix and label vector, we must encode our categorical features into binary variables. To do this, we will use the OneHotEncoder from the scikit-learn librarhy

```python
numerical_features = ['Age', 'RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak']
categorical_features = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']

encoder = OneHotEncoder(sparse_output=False)

encoded_array = encoder.fit_transform(df[categorical_features])

encoded_cols = encoder.get_feature_names_out(categorical_features)

encoded_df = pd.DataFrame(encoded_array, columns=encoded_cols)

X = pd.concat([df[numerical_features], encoded_df], axis=1)

X.head()
```

## Feature Scaling

Since KNN makes predictions by calculating distances between data points, we will need to scale our numberical features. We will StandardScaler() to accomplish this. This will ensure that all features contribute fairly, preventing features with large ranges from dominating the model. We have now finalized our feature matrix and can create our label vector as well.

```python
scaler = StandardScaler()
X_scaled = X.copy()
X_scaled[numerical_features] = scaler.fit_transform(X_scaled[numerical_features])
y = df['HeartDisease']

```

## Splitting into Train/Test Data

We will begin by splitting our data set with an 80/20 split and a random state of 42 using our normalized feature matrix. 

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state = 42, stratify=y) 
```

## Finding best K

Before evaluating our KNN model, we first needed to determine a suitable value for K, the number of nearest neighbors used in classification.
The choice of K controls the model’s complexity, smaller values can cause overfitting, while larger values may lead to underfitting by oversmoothing the decision boundary.

To identify a good value, we iteratively trained the KNN model using several K values and measured the resulting accuracy on the test set.

```python
k_accuracy = []
for k in [1, 3, 5, 11, 15, 27, 31, 35, 43, 46, 57, 56, 64, 79]:
    knn = KNeighborsClassifier(n_neighbors=k, weights='uniform')
    knn.fit(X_train, y_train)
    y_predict = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_predict)
    k_accuracy.append((k,accuracy))
    
    
for k in k_accuracy:
    print("Accuracy for k =", k[0], ":", k[1])
```

## Training our model

After testing various values of K, we have found that 11 is a good value to use. Now, we train our model using our testing data

```python
k = 11
knn = KNeighborsClassifier(n_neighbors=k, weights='uniform')
knn.fit(X_train, y_train)
y_predict = knn.predict(X_test)
y_proba = knn.predict_proba(X_test)[:, 1]
```

## Evaluating our model performance
We now evaluate the KNN model using standard classification metrics.  
These metrics show how accurately the model predicts heart disease cases in the test set.

We will compute:
- **Accuracy:** the proportion of total predictions that are correct  
- **Precision:** the proportion of positive predictions that are actually correct  
- **Recall:** the proportion of actual positive cases correctly identified  
- **F1-Score:** the harmonic mean of precision and recall, balancing both measures
- **ROC-AUC:** area under curve indicating how well the model distinguishes between positive and negative cases across classification

These values provide an overall sense of the model’s effectiveness and reliability.


```python
accuracy = accuracy_score(y_test, y_predict)
precision = precision_score(y_test, y_predict)
recall = recall_score(y_test, y_predict)
f1 = f1_score(y_test, y_predict)
roc_auc = roc_auc_score(y_test, y_proba)


results = pd.DataFrame({
    'Model': ['KNN'],
    'Accuracy': [accuracy],
    'Precision': [precision],
    'Recall': [recall],
    'F1-Score': [f1],
    'AUC': [roc_auc]
})

display(results.round(3))
```

## Training our model using Cross-Fold-Validation

To get a more reliable estimate of model performance, we evaluate our KNN model using 10-Fold Cross-Validation.
Instead of relying on one train/test split, this method splits the data into 10 equal folds. The model is trained and tested 10 times, each time using a different fold for testing.

This approach shows how consistent the model performs across different parts of the dataset and helps reduce the effect of one random split.
For each fold, we record accuracy, precision, recall, and F1-score, then average the results to get a more stable performance score.

```python
scoring = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']

cv_results = cross_validate(knn, X_scaled, y, cv=10, scoring=scoring)

cv_metrics = pd.DataFrame({
    'Fold': range(1, 11),
    'Accuracy': cv_results['test_accuracy'],
    'Precision': cv_results['test_precision'],
    'Recall': cv_results['test_recall'],
    'F1-Score': cv_results['test_f1'],
    'ROC-AUC': cv_results['test_roc_auc']
})

display(cv_metrics.round(3))

cv_summary = pd.DataFrame({
    'Model': ['KNN'],
    'Mean Accuracy': [np.mean(cv_results['test_accuracy'])],
    'Mean Precision': [np.mean(cv_results['test_precision'])],
    'Mean Recall': [np.mean(cv_results['test_recall'])],
    'Mean F1-Score': [np.mean(cv_results['test_f1'])],
    'Mean ROC-AUC': [np.mean(cv_results['test_roc_auc'])]
})

display(cv_summary.round(3).style.hide(axis="index"))
```

## Comparing our Results

Now that we have tried both evaluating methods, we can compare our results.

The comparison shows that the Train/Test Split produced slightly higher scores than the 10-Fold Cross-Validation. 

All metrics remain strong under both methods — with Accuracy (0.913 vs. 0.856), Precision (0.922 vs. 0.865), Recall (0.922 vs. 0.881), F1-Score (0.922 vs. 0.869), and ROC-AUC (0.952 vs. 0.913) showing consistent performance.
The difference between the two evaluations suggests that the KNN model performs well overall and generalizes effectively, even when tested more rigorously across multiple folds.

```python
comparison_df = pd.DataFrame({
    'Method': ['Train/Test Split', '10-Fold Cross-Validation'],
    'Accuracy': [results['Accuracy'][0], np.mean(cv_results['test_accuracy'])],
    'Precision': [results['Precision'][0], np.mean(cv_results['test_precision'])],
    'Recall': [results['Recall'][0], np.mean(cv_results['test_recall'])],
    'F1-Score': [results['F1-Score'][0], np.mean(cv_results['test_f1'])],
    'ROC-AUC': [results['AUC'][0], np.mean(cv_results['test_roc_auc'])],
})

display(comparison_df.round(3))


comparison_df.set_index('Method')[['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC']].plot(
    kind='bar', figsize=(8, 5), rot=0
)
plt.title('Comparison of KNN Evaluation Methods')
plt.ylabel('Score')
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Conclusion
In this notebook, we implemented and evaluated a K-Nearest Neighbors (KNN) model to predict heart disease using a mix of clinical and demographic features.

After encoding categorical variables and applying feature scaling, we tuned the number of neighbors (K) and evaluated model performance using both a standard 80/20 train–test split and 10-fold cross-validation.

The results show that KNN performs reliably, with high accuracy, precision, recall, F1-score, and ROC-AUC across both evaluation methods.
While the train/test split produced slightly higher results, cross-validation provided a more stable and realistic estimate of model performance on unseen data.

The best results were achieved with K = 11, which offered a balanced trade-off between bias and variance.
Smaller K values tended to overfit to the training data, while larger values smoothed the decision boundary too much and reduced accuracy.
The consistent cross-validation performance suggests that the model generalizes well without being overly sensitive to individual partitions.

Overall, KNN proves to be an effective model for this dataset.
