# K-Nearest Neighbors Classification - Heart Failure Prediction

This notebook applies K-Means Clustering to group patients with similar health characteristics in the Heart Disease dataset.
Unlike classification models such as Logistic Regression or K-Nearest Neighbors, K-Means is an unsupervised learning algorithm. It does not use labels like “heart disease” or “no heart disease”. Instead, it identifies natural groupings or patterns within the data based solely on the features provided.his notebook builds and evaluates a **K-Nearest Neighbors Classification** model for predicting the likelihood of heart failure based on patient health attributes.  

---

## Objective

Use K-Nearest Neighbors as a non-parametric classifier to:
- Predict the presence of heart disease based on similarity to other patients
- Compare its performance with models such as Logistic Regression, Decision Trees, and Random Forests
- Explore how the choice of k (number of neighbors) and distance metric affects prediction accuracy and model stability
---

## Dataset Summary

The dataset contains **918 patient records** and **11 health-related features**, along with a binary target variable:

- **Target:** `HeartDisease` (1 = patient has or is at risk of heart failure, 0 = otherwise)

**Features include:**
- *Age, Sex, ChestPainType, RestingBP, Cholesterol, FastingBS, RestingECG, MaxHR, ExerciseAngina, Oldpeak, ST_Slope*

---

## Overview

1. **Load the dataset** (dataset/heart.csv) 
2. **Preprocess features** (encode categorical variables and scale numerical features)  
3. **Train and evaluate** a KNN classifier using both a **standard 80/20 train–test split** and **10-fold cross-validation** 
4. **Compare performance** across the two evaluation methods to assess model stability and generalization  

---

## Notes

- All preprocessing (encoding, scaling, and splitting) is performed **within this notebook** for simplicity and reproducibility.  
- A fixed random seed (`random_state=42`) is used to maintain consistent splits and fair comparisons between methods.  

---


```python
# Gathering all neccessary libraries
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, auc
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import cross_validate
import matplotlib.pyplot as plt
```

## Reading our Dataset

We read our dataset from the Kaggle .csv file and and create our DataFrame

```python
df = pd.read_csv("dataset/heart.csv")

df.head()
```

```python
df.shape
```

```python
df.info()
```

## Encoding our categorical features

Before we create our feature matrix, we must encode our categorical features into binary variables. To do this, we will use the OneHotEncoder from the scikit-learn library

```python
numerical_features = ['Age', 'RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak']
categorical_features = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']

encoder = OneHotEncoder(sparse_output=False)

encoded_array = encoder.fit_transform(df[categorical_features])

encoded_cols = encoder.get_feature_names_out(categorical_features)

encoded_df = pd.DataFrame(encoded_array, columns=encoded_cols)

X = pd.concat([df[numerical_features], encoded_df], axis=1)

X.head()

```

## Feature Scaling


```python
scaler = StandardScaler()
X_scaled = X.copy()
X_scaled[numerical_features] = scaler.fit_transform(X_scaled[numerical_features])
```

## Fitting our model to normalized data

```python
my_KMeans = KMeans(n_clusters=3, random_state=42).fit(X_scaled)
labels  = my_KMeans.labels_
centers = my_KMeans.cluster_centers_
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
df["Clusters"] = labels

num_cols = ['Age','RestingBP','Cholesterol','FastingBS','MaxHR','Oldpeak']
cluster_means_original = df.groupby('Cluster')[num_cols].mean().round(2)
print(cluster_means_original)
```

```python

X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

x_col, y_col = 'Age', 'Cholesterol'
ix = X_scaled_df.columns.get_loc(x_col)
iy = X_scaled_df.columns.get_loc(y_col)

plt.figure(figsize=(8,6))
for l, c in zip([0,1,2], ['red','blue','green']):
    m = (labels == l)
    plt.scatter(X_scaled_df.iloc[m, ix],
                X_scaled_df.iloc[m, iy],
                color=c, marker='x', alpha=0.7, label=f'Cluster {l}')
    
    plt.plot(centers[l, ix], centers[l, iy], 'o',
             markerfacecolor=c, markeredgecolor='k', markersize=8)

plt.xlabel(f'{x_col} (scaled)')
plt.ylabel(f'{y_col} (scaled)')
plt.title('K-Means Clusters on Selected Features')
plt.xlim(-3, 3)
plt.ylim(-3, 3)  
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
